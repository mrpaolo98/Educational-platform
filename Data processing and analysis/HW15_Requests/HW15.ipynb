#### На главной страницы avito есть рекоммендации, необходимо сделать 1 http запрос на страницу, получить текст
#### и из данного текста достать большую часть рекомендаций
#### (для этого скорей всего Вам понадобится библиотека re - регулярные выражения)

```python
import requests
import re
```

```python
def extract_avito_urls() -> list:
    url = 'https://www.avito.ru/moskva'

    r = requests.get(url)
    res = []
    res_title = re.findall("""(?<=3R9dt">).*?(?=</span).*?(?<=1PdBw">).*?(?=</span)""", r.text)
    for item in res_title:
        title = re.findall(""".*?(?=</span)""", item)[0]
        price = re.findall("""(?<=1PdBw\">).*?(?=<!-- -->)""", item.replace(u'\xa0', u''))[0]
        res.append((title,price))
    return res
```

```python
data = extract_avito_urls()
```

```python
import pandas as pd

df = pd.DataFrame(data, columns=['Title', 'Price'])
df
```
