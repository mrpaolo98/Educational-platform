## Задача 4

1. Оцените распределение звонков зарубеж (какое количество какого числа звонков по выборке)
2. Посчитайте среднее значение числа звонков по выборке
3. Создайте новый признак с помощью apply, значение которого будет считаться как:

$$\frac{x - \bar{x}}{\sigma}$$

где  
$x$ - текущее значение звонков  
$\bar{x}$ - среднее значение звоноков за рубеж по выборке  
$\sigma$ - стандартное отклонение кол-ва звонков за рубеж по выборке  

4. Снова посчитайте среднее значение по выборке по новому фактору? Как оно изменилось? Почему?

```python
df.groupby(by=['Total intl calls'])['Total intl calls'].count().sort_values(ascending=False)
```

```python
mn = df['Total intl calls'].mean()
mn
```

```python
sd = df['Total intl calls'].std()
sd
```

```python
def nx(x):
    return (x - mn) / sd
```

```python
df['New column'] = df['Total intl calls'].apply(lambda x: nx(x))
```

```python
df
```

```python
df['New column'].mean() # среднее значение уменьшилось, так как применена формула стандартизованной оценки 
```
