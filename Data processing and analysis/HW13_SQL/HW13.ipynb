```python
# Создаем таблицы
cursor.execute("DROP TABLE IF EXISTS db_user")
cursor.execute("DROP TABLE IF EXISTS db_group")

cursor.execute("""

CREATE TABLE db_group (
    `id` INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
    `name` TEXT
)
""")

cursor.execute("""

CREATE TABLE db_user (
    `id` INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL ,
    `name` TEXT,
    `surname` TEXT,
    `age` INTEGER NOT NULL,
    `group_id` INTEGER NOT NULL,
    FOREIGN KEY (group_id) REFERENCES db_group(id)
)
""")
```

```python
# Обращаем внимание то что здесь немного другой синтаксис
add_user_query = """
INSERT INTO db_user (
    name, surname, age, group_id
) VALUES (
    ?, ?, ?, ?
)
"""
```

```python
add_group_query = """
INSERT INTO db_group (
    name
) VALUES (
    ?
)
"""
```

```python
group_names = (
    'Smart',
    'Happy',
    'Big',
    'Small',
    'Unly',
    'Beauty',
    'Hidden',
)
```

```python
# Добавляем данные в таблицы
for group_name in group_names:
    cursor.execute(add_group_query, (group_name,))
db_conn.commit()

for user in users:
    cursor.execute(add_user_query, tuple(user.values()))
db_conn.commit()
```

```python
import pandas as pd
```

```python
# Обращаем внимание на то что можно переименовывать колонки использую AS
group_df = pd.read_sql("""
SELECT 
    db_group.id AS id,
    db_group.name AS group_name
FROM db_group
""", db_conn, index_col='id')
group_df
```

```python
# Написать sql запрос который выводит следующий df
user_df = pd.read_sql("""
SELECT *
FROM db_user
""", db_conn, index_col='id'  
)
user_df
```

### Добавить в user_df group_name

```python
# Сделать merge который получает следующие данные(могут немного отличаться так как group_id рандомный)
user_with_group_df = user_df.merge(group_df, how='left', left_on='group_id', right_on='id')
user_with_group_df
```

```python
# Нужно написать sql запрос 
user_with_group_df_using_sql = pd.read_sql("""
SELECT 
    db_user.name AS name,
    db_user.surname AS surname,
    db_user.age As age,
    db_user.group_id, 
    db_group.name AS group_name
FROM db_user
INNER JOIN db_group
    ON db_user.group_id = db_group.id
""", db_conn  
)
user_with_group_df_using_sql
```

```python
# Для проверки - должно быть True
(user_with_group_df == user_with_group_df_using_sql).all().all() 
```

### Добавляем к group_df количество пользователей

```python
# Написать sql запрос который выводит следующее (мы не проходили group by поэтому по желанию задание)
group_with_count_users_df_using_sql = pd.read_sql("""
SELECT 
    db_group.id AS group_id,
    db_group.name AS group_name,
    COUNT(db_group.name) AS count_users
FROM db_user
INNER JOIN db_group
    ON db_user.group_id = db_group.id
GROUP BY group_id
""", db_conn
)
group_with_count_users_df_using_sql
```

```python
# Нужно сделать тоже самое но использую pandas
group_with_count_users_df = user_with_group_df.groupby(['group_id'])['group_name'].value_counts().reset_index(name='count_users')

group_with_count_users_df
```

```python
# Это для проверки - должно быть true
(group_with_count_users_df == group_with_count_users_df_using_sql).all().all()
```
