```python
class Point:
    
    def __init__(self, x=0, y=0, color='black'):
        self.x = x
        self.y = y
        self.color = color
        
    def __str__(self):
        return f'{__class__.__name__}(x={self.x}, y={self.y}, color={self.color})'
```

```python
point1 = Point(5, 10, 'red')
print(point1.x == 5)  # True
print(point1.y == 10)  # True
print(point1.color == 'red') # True
```

```python
point1 = Point(5, 10,)
print(point1.x == 5)  # True
print(point1.y == 10)  # True
print(point1.color == 'black') # True
```

```python
point1 = Point(5)
print(point1.x == 5)  # True
print(point1.y == 0)  # True
print(point1.color == 'black') # True
```

```python
point1 = Point()
print(point1.x == 0)  # True
print(point1.y == 0)  # True
print(point1.color == 'black') # True
```

### Реализовать методы add, sub

```python
class Point:
    
    def __init__(self, x=0, y=0, color='black'):
        self.x = x
        self.y = y
        self.color = color
        
    def move(self, a, b):
        self.x += a
        self.y += b
        
    def add(self, obj):
        return f'{__class__.__name__}(x={self.x + obj.x}, y={self.y + obj.y}, color=black)'
    
    def sub(self, obj):
        return f'{__class__.__name__}(x={self.x - obj.x}, y={self.y - obj.y}, color=black)'
        
    def __str__(self):
        return f'{__class__.__name__}(x={self.x}, y={self.y}, color={self.color})'
```

```python
point1 = Point(3, 5, 'red')
point2 = Point(4, 0, 'blue')
print(point1.add(point2)) # Point(x=7, y=5, color=black)
print(point1.sub(point2)) # Point(x=-1, y=5, color=black)
print(point1) # Point(x=3, y=5, color=red)
print(point2) # Point(x=4, y=0, color=blue)
```

## ***(не обязательно) сделать так что бы можно было выполнить следующий код

```python
class Point:
    
    def __init__(self, x=0, y=0, color='black'):
        self.x = x
        self.y = y
        self.color = color
        
    def move(self, a, b):
        self.x += a
        self.y += b
        
    def __add__(self, obj1):
        return Point(self.x + obj1.x, self.y + obj1.y, color = 'black') 
    
    def __sub__(self, obj1):
        return Point(self.x - obj1.x, self.y - obj1.y, color = 'black') 
        
    def __truediv__(self, k):
        return Point(int(self.x / k), int(self.y / k), color = 'black')
    
    def __mul__(self, k):
        return Point(self.x * k, self.y * k, color = 'black')
        
    def __str__(self):
        return f'{__class__.__name__}(x={self.x}, y={self.y}, color={self.color})'
```

```python
point1 = Point(3, 5, 'red')
point2 = Point(4, 0, 'blue')
point3 = point1 + point2
print(point3) # Point(x=7, y=5, color=black)
point4 = point1 - point2
print(point4) # Point(x=-1, y=5, color=black)
point5 = point1 * 5
print(point5) # Point(x=15, y=25, color=black)
point5 = point1 / 0.2
print(point5) # Point(x=15, y=25, color=black)
```
