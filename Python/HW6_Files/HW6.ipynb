## Создать функцию, которая считывает файл по имени в формате csv и возвращает список словарей

```python
import csv
```

```python
def read_f(f):
    data = []
    with open(f, 'r') as file_csv:
        reader = csv.reader(file_csv)
        for line in reader:
            data.append(line)
    my_list = []
    fieldnames = data[0]
    for values in data[1:]:
        res = dict(zip(fieldnames, values))
        res['Возраст'] = int(res['Возраст'])
        if res['Пол'] == 0:
            res['Пол'] = 'Мужской'
        else:
            res['Пол'] = 'Женский'
        my_list.append(res)
    return my_list
```

```python
def read_f_new(f):
    result = []
    with open(f, 'r') as file_csv:
        reader = csv.DictReader(file_csv)
        for line in reader:
            res = dict(line)
            res['Возраст'] = int(res['Возраст'])
            if res['Пол'] == 0:
                res['Пол'] = 'Мужской'
            else:
                res['Пол'] = 'Женский'
            result.append(res)
    return result
```

## Написать функцию, которая принимает название файл (hw.json) и возвращает средний возраст(27) и вызывать её

```python
def read_json(f):
    with open(f, 'r') as file_json:
        data = file_json.read()
        lst1 = json.loads(data)
    return sum([item['Возраст'] for item in lst1])//len(lst1)
```

```python
print(read_json('hw.json'))
```
